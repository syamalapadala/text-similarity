```python
from sentence_transformers import SentenceTransformer, util
import torch
import pandas as pd
```

```python
model=SentenceTransformer("sentence-transformers/all-MiniLM-l6-v2")
```

```python
df=pd.read_csv("/content/test_changes1.csv")
```

```python
df
```

```python
def get_embeddings(model,df,col_name):
  list_companies=list(df[col_name])
  list_embeddings=[]
  for i in list_companies:
    l=model.encode(i,convert_to_tensor=True)
    list_embeddings.append(l)
  return list_embeddings
```

```python
list_embeddings=get_embeddings(model,df,"company")
```

```python
def get_similarity(query,embeddings,model,list_values):
  query_vec=model.encode(query,convert_to_tensor=True)
  similarity_matrix=[]
  for j in embeddings:
    similarity_matrix.append(float(util.pytorch_cos_sim(query_vec,j)[0][0]))
  print(similarity_matrix)
  index=find_largest_position(similarity_matrix)
  return list_values[index]
```

```python
def find_largest_position(lst):
    filtered_list = [(idx, val) for idx, val in enumerate(lst) if val != 1]
    if filtered_list:
        return max(filtered_list, key=lambda x: x[1])[0]
    else:
        return None
```

```python
list_values=list(df["company"])
```

```python
get_similarity("google",list_embeddings,model,list_values)
```
